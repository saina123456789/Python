```python
def binning(data):
    maximum_value=max(data)
    minimum_value=min(data)
    print(maximum_value,minimum_value)
    #child,teen,adult,old
    child=[]#empty bin/list
    teen=[]
    adult=[]
    old=[]
    age=[10,1]
#for in range(0,len(data)):
    for i in data:
        if i<=12:
            child.append(i)#insert value in ur list
        elif i >12 and i<20:
            teen.append(i)
        elif i>=20 and i<45:
            adult.append(i)
        elif i>45:
            old.append(i)
    print("Bin of children=",child)
    print("Bin of teen=",teen)
    print("Bin of adult=",adult)
    print("Bin of old=",old)
        
age=[10,1,30,15,86,3,]
binning(age)
                    
        
```
